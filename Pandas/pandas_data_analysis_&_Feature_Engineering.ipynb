# Feature Engineering


```python
#Feature engineering
df = pd.read_csv("raw_data.csv")

df2 = df.copy()
df2["tax"] = df2["income"].apply(lambda x : "20%" if x>=60000 else "10%")

gender_map = {"Male" : "M", "Female":"F","Unknown": "U"}
df2["gender"] = df2["gender"].map(gender_map)

df2 = df2.assign(new_income = df2["income"] * 1.1)

df2["country"] = df2["country"].replace("USA", "US")
df2.columns = ["Id", "Name","Age","Country","Gender","Income","Tax","New_Income"]
df2 = df2.rename(columns={"Income" : "Salary"})
#df2 = df2.rename(index={2:"First"})

df2.sort_values("Salary")
df2 = df2.fillna(50)
#df2.sort_values("Income", ascending = False)


sorted_df = df2.sort_values(["Salary","Age"]) 

#sorted_df = df2.sort_values(["Salary"],ascending = False)
sorted_df.sort_index()

# #reset
sorted_df.reset_index()
sorted_df.reset_index(drop = True)

# #ranking

sorted_df["Ranking"] = sorted_df["Salary"].rank(ascending = False , method = "max")
sorted_df[["Id", "Name","Age","Country","Gender","Salary","New_Income","Tax"]]


sorted_df
#df2
```

# Grouping And Aggregation Of Data

```python
df.groupby("country")["income"].mean()
df.groupby("country")["income"].min()
df.groupby("country")["income"].max()

df.groupby("gender")["income"].mean()
df.groupby("gender")["income"].min()
df.groupby("gender")["income"].max()

df.groupby("country")["income"].agg(["mean","min","max"])

df.groupby("country")["income"].agg(avg_sal = "mean",min_sal = "min",max_sal = "max")

df.groupby("country").agg({
    "income" :"mean",
    "age" : "min"
})
df.groupby("country").agg(
    max_salary = ("income","max"),
    avg_age = ("age" ,"mean")
)
```

# Melt and Pivot for RESHAPING


```python
#Melt and Pivot for RESHAPING

df = pd.DataFrame({
    "country":["USA","USA","India","India"],
    "year": [2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22],
   
})


melted_df = df.melt(
    id_vars = ["country","year"],
    value_vars = ["sales","profit"],
    var_name = "metrics",
    value_name = "value"
)


melted_df.pivot(
    index = ["country", "year"],
    columns ="metrics",
    values = "value"
)    
```

# Merging and Joining Data

```python
df_customers = pd.DataFrame({
    "customer_id" : [1,2,3,4],
    "name" : ["Adam", "Bob", "Charlie","Dave"]

})

df_orders = pd.DataFrame({
    "order_id" : [101, 102, 103, 104],
    "customer_id": [2,1,4,5],
    "amount":[250,120,300,180]
})

pd.merge(df_customers, df_orders, on = "customer_id") #inner join (matching values)
pd.merge(df_customers, df_orders, on="customer_id", how="left") #left join 
pd.merge(df_customers, df_orders, on="customer_id", how="right") #right join 
pd.merge(df_customers, df_orders, on="customer_id", how="outer") #outerjoin (valid values of both tables)

```

# Concatenation
